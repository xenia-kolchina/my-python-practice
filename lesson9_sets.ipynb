```python
# Задача 1. Создайте список из всех натуральных чисел от 1 до 50 включительно, используя функцию range. 
# Превратите этот список в множество. Добавьте числа 21, 24, 29, 53, 104. Чему равна len этого множества?
x = []
for i in range(1,51,1):
    x.append(i)
setx = set(x)
setx.update([21,24,29,53,104])
len(setx)
```

```python
# Задача 2. Создайте список из всех натуральных чисел от 1 до 300 включительно. 
# Затем создайте список из целых четных чисел от 115 до 229. Для удобства используйте функцию range.
# Превратите эти списки в множества. 
# Создайте третье множество, которое будет объединением первых двух. 
# Полученное множество переведите в список и посчитайте сумму. 
# Подсказка: Чтобы объединить множества, используйте метод union.
x = set(range(1,301))
y = set(range(115,230))
z = x.union(y)
z = list(z)
sum(z)
```

```python
# Задача 3. Дано два множества:
# x={25,33,67,89}
# y={26,33,67,90,101}
# Какой метод нужно применить, чтобы получить следующее множество?
# y={26, 101, 89, 90, 25}
# В ответе запишите название без точек и скобок
x={25, 33, 67, 89}
y={26, 33, 67, 90, 101}
print(x.symmetric_difference(y))
```

```python
# Задача 4. Создайте список из чисел от 100 до 300 включительно, которые делятся на 17.
# Для удобства используйте функцию range и условие. Превратите список в множество set_1.
# Создайте второе множество set_2 из всех целых чисел от 200 до 400 включительно.
# Найдите множество чисел set_3, которые входят в set_1, но не входят во второе множество set_2.
# Чему равна длина множества set_3?
# Подсказка: Вам могут пригодиться append() и difference().
x = []
for i in range(100,301):
    if i % 17 == 0:
        x.append(i)
set_1 = set(x)
set_2 = set(range(200,401))
set_3 = set_1.difference(set_2)
len(set_3)
```

```python
# Задача 5. Дано два сета.
# set_1 = {97, 65, 100, 6, 39, 43, 46, 55, 57, 29}
# set_2 = {33, 98, 97, 4, 37, 36, 6, 34, 59, 75, 77, 47, 83, 84, 51, 91, 92, 30}
# Оставьте первый сет set_1 без изменения, а во втором сете set_2 оставьте только те значения, которые делятся на 2.
# После сделайте новый сет set_3, который содержит значения, которые есть только в set_1 или set_2 , но не в двух одновременно.
# В ответ запишите сумму всех значений нового сета.

# Подсказка: Чтобы убрать из второго множества set_2 числа, которые не делятся на 2, сделайте следующее:
# Создайте новый пустой список list_2.
# Преобразуйте множество set_2 в список.
# Пройдитесь по всем элементам списка set_2, добавляя в новый список list_2 только те элементы, которые делятся на 2.
# Преобразуйте list_2 в множество.

set_1 = {97, 65, 100, 6, 39, 43, 46, 55, 57, 29}
set_2 = {33, 98, 97, 4, 37, 36, 6, 34, 59, 75, 77, 47, 83, 84, 51, 91, 92, 30}
x = []
for i in set_2:
    if i % 2 == 0:
        x.append(i)
set_2 = set(x)
set_3 = set_1.symmetric_difference(set_2)
sum(set_3)
```
