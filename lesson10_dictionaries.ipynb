```python
# Пусть у нас есть словарь и два списка:
# dict4 = {1:1, 2:4, 4:16}
# list4 = [3, 4, 6]
# list5 = [2, 1, 0]

# Выполним следующее преобразование:
# Посчитаем сумму элементов с одним индексом в двух списках;
# Найдем квадрат каждой суммы;
# Добавим в словарь каждую сумму как ключ, а ее квадрат - как значение.

dict4 = {1:1, 2:4, 4:16}
list4 = [3, 4, 6]
list5 = [2, 1, 0]
list6 = [list4[i]+list5[i] for i in range(len(list4))]
list7 = [j**2 for j in list6]
for k in range(len(list6)):
    dict4.update({list6[k]:list7[k]})
dict4
```

```python
# Допустим, у нас есть пустой словарь и два списка:
# dict5 = {}
# list4 = [3, 4, 6]
# list5 = [2, 1, 0]

# Выполним следующие преобразования:

# Посчитаем сумму элементов с одним индексом в двух списках;

# Посчитаем разность элементов с одним индексом в двух списках;

# Добавим в словарь сумму как ключ, а соответствующую разность - как значение;

# При этом если сумма элементов повторяется, то мы используем только первый 
# ключ (первую посчитанную разность), последующие игнорируем.

# Подсказка: воспользуйтесь методом get()

dict5 = {}
list4 = [3, 4, 6]
list5 = [2, 1, 0]
list6 = [list4[i]+list5[i] for i in range(len(list4))]
list7 = [list4[i]-list5[i] for i in range(len(list4))]
for k in range(len(list6)):
    if dict5.get(list6[k])==None:
        dict5.update({list6[k]:list7[k]})
dict5
```

```python
# Пусть у нас есть словарь:

# dict6 = {'яблоко': 3, 'банан': 2, 'апельсин': 4, 'груша': 1}
# Исключим из него пары со значениями меньше 3.

# Подсказка: используйте цикл for key in dict.keys() для перебора всех ключей словаря.

dict6 = {'яблоко': 3, 'банан': 2, 'апельсин': 4, 'груша': 1}

# list(dict6.keys()) создает отдельный список, который не меняется
for key in list(dict6.keys()): 
    if dict6[key] < 3:
        dict6.pop(key)

print(dict6)
```

```python
# У нас есть словарь и два списка:
# a = {1:1, 2:4, 3:9, 4:16}
# b = [4,8,5]
# c = [2,3,1]

# Выполним следующие преобразования:

# Посчитаем разность элементов с одним индексом в двух списках;

# Если разность совпадает с одним из ключей словаря, то удалите пару 
# ключ:значение с таким ключом.

# Подсказка: используйте метод pop(разность, None) для удаления пары 
# ключ:значение.

a = {1:1, 2:4, 3:9, 4:16}
b = [4,8,5]
c = [2,3,1]
d = [b[i]-c[i] for i in range(len(b))]
for key in list(a.keys()):
    if key in d:
        a.pop(key,None)
a
```

```python
# Есть два словаря с продажами сотрудников: за месяц monthly_sales и за один день daily_sales.
# Требуется добавить к словарю с месячными продажами monthly_sales продажи сотрудников за день.
# Примечание: в словаре daily_sales есть новые сотрудники, их нужно добавить, как новую пару ключ:значение, 
# а для тех сотрудников, которые уже есть в исходном словаре monthly_sales нужно просто прибавить их новую продажу.
# В ответ укажите длину нового словаря.
# Подсказка: Используйте цикл, условие и метод items().

monthly_sales = {
    'Иван Иванов': 15000,
    'Петр Петров': 12000,
    'Сидор Сидоров': 8000,
    'Ольга Иванова': 10000,
    'Антон Петров': 9000,
    'Екатерина Сидорова': 11000,
    'Андрей Иванов': 13000,
    'Мария Петрова': 7000,
    'Дмитрий Сидоров': 12500,
    'Светлана Иванова': 9500,
    'Алексей Петров': 10500
}

daily_sales = {
    'Петр Петров': 1800,
    'Николай Иванов': 3000,
    'Сергей Петров': 2200,
    'Евгений Сидоров': 1500,
    'Екатерина Сидорова': 1000,
    'Василий Иванов': 2800,
    'Павел Петров': 1700,
    'Максим Сидоров': 2100,
    'Анна Иванова': 1900,
    'Олег Петров': 2300,
    'Денис Сидоров': 1600
}

# Обновление словаря с месячными продажами
for employee, daily_sale in daily_sales.items():
    if employee in monthly_sales:
        monthly_sales[employee] += daily_sale
    else:
        monthly_sales[employee] = daily_sale

# Вывод длины нового словаря
print(len(monthly_sales))
```

```python
# Используйте обновленный список monthly_sales с продажами сотрудников за месяц и за последний день, 
# и добавьте каждому сотруднику бонусы за продажи.

# Бонусы начисляются исходя из суммы продаж:
# До 10 000 рублей: Бонус не начисляется.
# От 10 000 (включительно) до 15 000 (не включительно) рублей: Бонус составляет 5% от суммы продаж.
# От 15 000 до 20 000 рублей: Бонус составляет 10% от суммы продаж.
# Свыше 20 000 рублей: Бонус составляет 15% от суммы продаж.

# Обратите внимание: Нижняя граница берется включительно, а верхняя - нет!
# Подсчитайте общую сумму бонусов, которые компания должна заплатить всем сотрудникам.
# В ответ напишите число - общую сумму бонусов.

# Подсказка: Используйте цикл, условие (if - elif) и метод items().

bonus = 0
for zp in list(monthly_sales.values()):
    if zp < 10000:
        bonus += 0
    elif zp >= 10000 and zp < 15000:
        bonus += zp*0.05
    elif zp >= 15000 and zp < 20000:
        bonus += zp*0.1
    else:
        bonus += zp*0.15
bonus
```
