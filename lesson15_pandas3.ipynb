```python
# Давайте вернёмся к оригинальному датасету за 2019 г. В этом задании работайте с оригинальными данными, без преобразований, 
# выполненных в задачах 1 и 2.

# Создайте новый столбец:
# Если название страны состоит из одного слова, то в новом столбце должно быть записано '1 слово в названии'.
# Если название страны состоит из двух и более слов, то в новом столбце должно быть записано '2 и более слов в названии'.

# Затем сгруппируйте данные по новому столбцу и вычислите среднее значение Score. 
# Ответ округлите до 2 знаков после запятой. В ответе укажите значение для стран с 2 и более словами в названии.
df = pd.read_csv(path) 

df['country_split']=df['Country or region'].apply(lambda x: x.split(' '))
df['len_country']=''
for i in range(len(df)):
    if len(df['country_split'][i])==1:
        df['len_country'][i]='1 слово в названии'
    if len(df['country_split'][i])>1:
        df['len_country'][i]='2 и более слов в названии'
df_gr=df.groupby('len_country')['Score'].agg('mean')
df_gr=df_gr.round(2)
display(df_gr.head(10))
```
